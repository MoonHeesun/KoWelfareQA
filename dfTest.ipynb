```python
import re
```

```python
def strip_replace_text(s: str):
	regex = '([^가-힣0-9a-zA-Z.,·•%↓()\s\\\])'
	s = re.sub(pattern=regex, repl="", string=s)
	s = s.replace(' ', '')
	s = s.replace('•','·')
	s = s.replace('Ⅰ','')
	s = s.replace('Ⅱ','')
	return s

# def key_process(s: str):
# 	s = s.strip()
# 	s = s.replace('\n', '')
# 	s = s.replace(' ', '')
# 	s = s.replace('•', '·')
# 	s = s.replace("#", '')
# 	s = s.replace("‘", '')
# 	s = s.replace("’", '')
	
# 	return s
```
